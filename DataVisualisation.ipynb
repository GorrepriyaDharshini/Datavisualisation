```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
titanic_df=pd.read_csv("C:/Users/priyadharshini gorre/Downloads/Datasets-main/Datasets-main/titanic-training-data.csv")
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Create countplot of passangers by class
sns.countplot(x='Pclass', data=titanic_df)

#set the title and axis labels
plt.title('Passangers by Class')
plt.xlabel('Class')
plt.ylabel('Count')

# show the plot
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='Pclass',y='Age',data=titanic_df)

plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass',y='Age',data=titanic_df)

plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')

plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Create swarmplot of passangers by class
sns.swarmplot(x='Pclass', data=titanic_df)

#set the title and axis labels
plt.title('Passangers by Class')
plt.xlabel('Class')
plt.ylabel('Count')

# show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Create countplot of passangers by class
sns.countplot(x='Pclass',hue='Sex', data=titanic_df)

#set the title and axis labels
plt.title('Passangers by Class and Sex')
plt.xlabel('Class')
plt.ylabel('Count and Sex')

# show the plot
plt.show()
```

```python
corr=titanic_df.corr()
```

```python
# Set the figure size
plt.figure(figuresize=(8, 6))

# Create countplot of passangers by class
sns.heatmap(corr,annot=True)

#set the title and axis labels
plt.title('Correalation Matrix')
plt.xlabel('Features')
plt.yabel('Features')

plt.show()
```
